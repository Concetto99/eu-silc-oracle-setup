### OUTPUT THE SCRIPT TO CREATE TABLES

```python
datasets = (UDB_cIT23D, UDB_cIT23H, UDB_cIT23P, UDB_cIT23R)
datasets_str = ("UDB_cIT23D", "UDB_cIT23H", "UDB_cIT23P", "UDB_cIT23R")
db_user = "C##EUSILC_SVIL" # Set with your user

for d in range(0,len(datasets)):

    column_vectors = datasets[d].columns

    print(f"CREATE TABLE {db_user}.TD_{datasets_str[d]}", "(")
    for i in range(0,len(column_vectors)):

        data_type = datasets[d][column_vectors[i]].dtypes
        if data_type == "int64":
            data_type = "NUMBER(38)"
        elif data_type == "float64":
            data_type = "NUMBER(38,12)"
        else: data_type = "VARCHAR2(100)"
        if i == 0:
            print(column_vectors[i], " ", data_type , " ")
        else:
            print(",", column_vectors[i], " ", data_type , " ")
    print(");\n\n")

```
